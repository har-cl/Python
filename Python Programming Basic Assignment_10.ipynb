```python
print('Write a Python program to find sum of elements in list?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)
sum=0
for i in l:
    sum+=i
print('sum of all element is : ',sum)

print('Write a Python program to Multiply all numbers in the list?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)
multi=1
for i in l:
    multi*=i
print('multiplication of all element is : ',multi)

print('Write a Python program to find smallest number in a list?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)
small=l[0]
for i in l:
    if i<small:
        small=i
print('smallest of all element is : ',small)

print('Write a Python program to find largest number in a list?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)
large=l[0]
for i in l:
    if i>large:
        large=i
print('largest of all element is : ',large)

print('Write a Python program to find second largest number in a list?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)
large=l[0]
sl=l[1]
for i in l:
    if i>large:
        sl=large
        large=i
print('second largest of all element is : ',sl)

print('Write a Python program to find N largest elements from a list?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)

l.sort()
l=l[::-1]
n=int(input('Enter N for Nth largest : '))
print('Nth largest element is : ',l[n-1])

print('Write a Python program to print even numbers in a list?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)
print('Even numbers are : ')
for i in l:
    if i%2==0:
        print(i)
        
print('Write a Python program to print odd numbers in a List?')

n=int(input('Enter no of elements in array : '))
l=[]
for i in range(0,n):
    e=int(input('enter element : '))
    l.append(e)
print('Entered array is : ',l)
print('Odd numbers are : ')
for i in l:
    if i%2==1:
        print(i)
        
print('Write a Python program to Remove empty List from List?')

l=[434,'harshil',[234,34,35],[],45,[],'chauhan']
print('Entered list is : ',l)

for i in l:
    if type(i)==list:
        if len(i)<1:
            l.pop(l.index(i))
print('By removing empty list, final list is : ',l)

print('Write a Python program to Cloning or Copying a list?')

l=[434,'harshil',[234,34,35],[],45,[],'chauhan']

def copy_list(l):
    l_copy=l
    return l_copy

new_list=copy_list(l)
print('Orignal list is : ',l)
print('Copied list is : ',new_list)

print('Write a Python program to Count occurrences of an element in a list?')

l=[434,'harshil',[234,34,35],[],45,[],'chauhan']
count=0

for i in l:
    count+=1
    
print('No. of elements are : ',count)
```
