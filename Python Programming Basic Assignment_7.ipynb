```python
print('Write a Python Program to find sum of array?')
a=int(input('Enter no. of elements: '))
l=[]
sum=0
for i in range(0,a):
    n=int(input('Enter value: '))
    sum+=n
    l.append(n)
print('Array is : ',l)
print('sum of array is : ',sum)

print('Write a Python Program to find largest element in an array?')
a=int(input('Enter no. of elements: '))
l=[]
max=0
for i in range(0,a):
    n=int(input('Enter value: '))
    if n>max:
        max=n
    l.append(n)
print('Array is : ',l)
print('Max of array is : ',max)

print('Write a Python Program for array rotation?')
a=int(input('Enter no. of elements: '))
l=[]
rot=int(input('enter value of rotation : '))
for i in range(0,a):
    n=int(input('Enter value: '))
    l.append(n)
l1=[]
l2=[]
lt=l*2
for i in range(0,len(l)):
    l1.append(lt[rot+i])
for i in range(0,len(l)):
    l2.append(lt[len(l)-rot+i])   
print('Inserted array is : ',l)
print('Left rotated array is : ',l1)
print('Right rotated array is : ',l2)

print('Write a Python Program to Split the array and add the first part to the end?')
a=int(input('Enter no. of elements: '))
l=[]
split=int(input('enter index of after which split is needed : '))
for i in range(0,a):
    n=int(input('Enter value: '))
    l.append(n)
l1=[]
lt=l*2
for i in range(0,len(l)):
    l1.append(lt[split+i])  
print('Inserted array is : ',l)
print('splited array is : ',l1)

print('Write a Python Program to check if given array is Monotonic?')
a=int(input('Enter no. of elements: '))
l=[]
for i in range(0,a):
    n=int(input('Enter value: '))
    l.append(n)
mono=0
mi=True
md=True
for i in range(0,len(l)-1):
    if l[i+1]<l[i]:
        mi=False            
for i in range(0,len(l)-1):
    if l[i+1]>l[i]:
        md=False
mono=mi or md
print('Inserted array is : ',l)
if mono==True:
    print('Entered function is monotonic')
else:
    print('Entered function is not monotonic')
```
